```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
raw_data = pd.read_csv('2.02. Binary predictors.csv')
raw_data
```

```python
data = raw_data.copy()
data['Admitted'] = data['Admitted'].map({'Yes':1, 'No':0})
data['Gender'] = data['Gender'].map({'Female':1, 'Male':0})
data
```

## Gender and Admittance

```python
y = data['Admitted']
x1 = data['Gender']
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
results_log.summary()
```

## SAT, Gender and Admittance

```python
y = data['Admitted']
x1 = data[['SAT', 'Gender']]
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
results_log.summary()
```

```python
#The log-likelihood increased compared to when only Gender was included >>> model improved
#SAT was more outstanding predictor(the p-value for Gender was 0, but it changed to 0 for SAT, indicating the change)
```

```python
test_actual = test['Admitted']
test_data = test.drop(['Admitted'], axis=1)
test_data = sm.add_constant(test_data)
#test_data = test_data[x.columns.values]
test_data
```

```python
#sm.LogitResults.pred_table() does not provide testing as a functionality unlike sklearn
#manually create a function of confusion matrix
```

There will be three arguments, the input data, the actual values.
Our function will use the already created regression model to make predictions based on the data. Then it will summarize the values in a table.
Finally, make it calculate the accuracy too. Once used, the function will return the confusion matrix and the accuracy. (Detailed comments는 resources 파일 참고)

```python
def confusion_matrix(data, actual_values, model):

        pred_values = model.predict(data)
        bins=np.array([0,0.5,1])
        cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
        accuracy = (cm[0,0]+cm[1,1])/cm.sum()
        return cm, accuracy
```

```python
cm = confusion_matrix(test_data, test_actual, results_log)
cm
```

```python
#The test accuracy is the figure we use when we refer to overall accuracy
```
