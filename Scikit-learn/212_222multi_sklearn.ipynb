# Formula for Adjusted R^2

$R^2_{adj.} = 1 - (1-R^2)*\frac{n-1}{n-p-1}$

```python
#n = the number of observations
#p = the number of predictors
```

```python
x.shape
```

```python
r2 = reg.score(x,y)

n = x.shape[0]
#n=84지만 x.shape 형태로 parameterize 해줌
p = x.shape[1]
#p=2지만 x.shape 형태로 parameterize 해줌

adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```

# Feature Selection

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(x,y)
```

```python
p_values = f_regression(x,y)[1]
p_values
```

```python
p_values.round(3)
```

```python
#reg_summary = pd.DataFrame(data=['SAT', 'Rand 1,2,3'], columns=['Features'])를 parameterize
#이게 왜 필요하지? 나중에 변수가 n개가 되었을 때 편리성을 위해서!
reg_summary = pd.DataFrame(data=x.columns.values, columns=['Features'])
reg_summary
```

```python
reg_summary['Coefficients'] = reg.coef_
reg_summary['p-values'] = p_values.round(3)
```

```python
reg_summary
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
#we have created an empty StandScaler object
#'scaler' will be used to subtract the mean and divide by the standard deviation
```

```python
scaler.fit(x)
```

```python
#Scaling Mechanisms
##StandardScaler(copy=Trye, with_mean=True, with_std=True)
##'fit' calculates and stores the mean and standard deviation of each feature
##scaler contains all standardization information
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

# Regression with scalred features

```python
reg = LinearRegression()
reg.fit(x_scaled, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg_summary = pd.DataFrame([['Intercept'],['SAT'],['Rand 1,2,3']], columns=['Features'])
reg_summary['Weights'] = reg.intercept_, reg.coef_[0], reg.coef_[1]
```

```python
reg_summary
```

```python
reg_summary = pd.DataFrame([['Bias'],['SAT'],['Rand 1,2,3']], columns=['Features'])
reg_summary['Weights'] = reg.intercept_, reg.coef_[0], reg.coef_[1]
#machine learning 용어들로 바꿔줘서('intercept'-> 'Bias' convenient summary table과 new one을 구분함
```

```python
reg_summary
```
