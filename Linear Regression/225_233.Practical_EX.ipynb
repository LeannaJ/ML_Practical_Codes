## Linear regression model

### Declare the inputs and the targets

```python
#Our target is the log price
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

### Scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

```python
#not usually recommended to standardize dummy variables
#To scale only the continuous variables, refer to the logistics regression part
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

### Create the regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
# It is not simple linear regression
# In fact, this is a log-linear regression as the dependent variable is the log of 'Price'
```

```python
# 1) To plot the predicted values against the observed values
# A simple way to check the final results
```

```python
y_hat = reg.predict(x_train)
# B/C in multiple regression equation, y hat is inferred value
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel("Targets (y_train)",size=18)
plt.ylabel("Predictions (y_hat)",size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
# 2) To use residual plot
# residual = differences btw the targets and the predictions
```

```python
sns.displot(y_train - y_hat)
plt.title("Residuals PDF", size=18)
```

```python
#residuals are the estimates of the errors >>> normality assumption OK
#There are certain observations for which (y_train-y_hat) is much lower than the mean
#(a much higher price is predicted than is observed)

#We can conclude that the predictions rarely underestimate the targets
```

```python
#r-squared
reg.score(x_train,y_train)
```

### measuring performance of the predictions

```python
df_pf = pd.DataFrame(y_hat_test, columns=['Prediction'])
df_pf.head()
```

```python
#This prediction is not simple prediction, but log prediction
#Need to take the exponentials of the logs to revert to the original price prediction

df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
#Why are there so many missing values randomly? Let’s check the y_test
y_test
```

```python
#After checking, it turns out there are indexes present
#When performing train_test_split, the original indexes are preserved
#To avoid missing values when adding y_test to df_pf, remove the original indexes to ensure they match properly

#Overwrite y_test with y_test reset index with an argument drops equals true
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
#According to the OLS framework, “residual” is a more accurate term than “difference” (minimize SSE)
#lower error means better explanatory power
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```
