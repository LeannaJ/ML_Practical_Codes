```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set_theme()
```

```python
raw_data = pd.read_csv('1.03. Dummies.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1,'No':0})
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'],y)
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='#006837')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='#a50026')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show() 
```

```python
plt.scatter(data['SAT'],y,c=data['Attendance'],cmap='RdYlGn_r')
#Depending on the value of data['Attendance'], the color of each point changes
#cmap='RdYlGn_r' specifies the color map, using the ‘RdYlGn_r’ (red-yellow-green reversed) color map
#Higher data['Attendance'] values are displayed in green, while lower values are displayed in red
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='#006837')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='#a50026')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
plt.scatter(data['SAT'],y,c=data['Attendance'],cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
yhat = 0.275 + 0.0017*data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='#006837', label = 'regression line1')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='#a50026', label = 'regression line2')
fig = plt.plot(data['SAT'], yhat, lw=2, c='#4C72B0', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
#Bob, who got 1700 on the SAT and did Not attend
#Alice, who got 1670 on the SAT and attended
new_data = pd.DataFrame({'const':1, 'SAT':[1700,1670], 'Attendance':[0,1]})
new_data = new_data[['const','SAT','Attendance']]
new_data
```

```python
new_data.rename(index={0:'Bob', 1:'Alice'})
```

```python
#The fitted regression for us is in the variablr: results
#results = sm.OLS(y,x),fit()
predictions = results.predict(new_data)
predictions
```

```python
#Convert the above data into a pd.DataFrame format and join it with the new_data table
predictionsdf = pd.DataFrame({'Predictions': predictions})
joined = new_data.join(predictionsdf)
joined.rename(index={0: 'Bob', 1: 'Alice'})
```
